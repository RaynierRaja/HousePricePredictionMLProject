```python
import os
import tarfile
from six.moves import urllib
```

```python
url = "https://github.com/ageron/handson-ml/blob/master/datasets/housing/housing.tgz?raw=true"
saveDir = "datasets/housing"
saveFileName = "housing.tgz"
```

```python
def downloadData(url,saveDir,saveFileName):
    # Create Directory
    if not os.path.isdir(saveDir):
        os.makedirs(saveDir)
    # Download
    tgzPath = os.path.join(saveDir,saveFileName)
    urllib.request.urlretrieve(url,tgzPath)
    # Extract
    tgzData = tarfile.open(tgzPath)
    tgzData.extractall(path = saveDir)
    tgzData.close()    
```

```python
downloadData(url,saveDir,saveFileName)
```
